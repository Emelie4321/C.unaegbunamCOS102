```python
import numpy as np

def get_roots(coefficients):
    """Calculates roots for any polynomial given its coefficients."""
    try:
        roots = np.roots(coefficients)
        return roots
    except Exception as e:
        return f"Error calculating roots: {e}"

def main():
    print("--- Polynomial Root Finder ---")
    
    while True:
        print("\nWhat operation would you like to perform?")
        print("1. Find roots of a Quadratic Equation (Ax² + Bx + C = 0)")
        print("2. Find roots of a Cubic Equation (Ax³ + Bx² + Cx + D = 0)")
        print("3. Find roots of a Quartic Equation (Ax⁴ + Bx³ + Cx² + Dx + E = 0)")
        print("4. Exit")
        
        choice = input("\nEnter choice (1-4): ")
        
        if choice == '4':
            print("Exiting program. Goodbye!")
            break
            
        coeffs = []
        try:
            if choice == '1':
                a = float(input("Enter A: "))
                b = float(input("Enter B: "))
                c = float(input("Enter C: "))
                coeffs = [a, b, c]
            elif choice == '2':
                a = float(input("Enter A: "))
                b = float(input("Enter B: "))
                c = float(input("Enter C: "))
                d = float(input("Enter D: "))
                coeffs = [a, b, c, d]
            elif choice == '3':
                a = float(input("Enter A: "))
                b = float(input("Enter B: "))
                c = float(input("Enter C: "))
                d = float(input("Enter D: "))
                e = float(input("Enter E: "))
                coeffs = [a, b, c, d, e]
            else:
                print("Invalid selection. Please try again.")
                continue

            # Calculate roots
            results = get_roots(coeffs)
            
            print("\nThe roots are:")
            for i, r in enumerate(results, 1):
                # Clean up display: round and handle complex formatting
                real = round(r.real, 4)
                imag = round(r.imag, 4)
                if imag == 0:
                    print(f"  Root {i}: {real}")
                else:
                    sign = "+" if imag > 0 else "-"
                    print(f"  Root {i}: {real} {sign} {abs(imag)}j")
                    
        except ValueError:
            print("Invalid input! Please enter numerical values for coefficients.")

if __name__ == "__main__":
    main()
```
