BEGIN
    // Customer selects items and adds them to cart

    // Initialize total cost to zero
    totalCost = 0

    // Loop through each item in the cart
    FOR EACH item IN cart DO
        // Get the price of the item
        itemPrice = item.price

        // Add the item price to the total cost
        totalCost = totalCost + itemPrice
    END FOR

    // Apply any discounts
    IF discountExists THEN
        totalCost = totalCost - discountAmount
    END IF

    // Calculate sales tax
    salesTax = totalCost * taxRate
    finalCost = totalCost + salesTax

    // Display the final cost to the customer
    DISPLAY finalCost
    // Ask the customer for payment
    DISPLAY "How would you like to pay?"
    INPUT paymentMethod

    // Process payment based on the payment method
    IF paymentMethod == "cash" THEN
        // Ask for cash amount
        DISPLAY "Enter cash amount:"
        INPUT cashAmount

        // Calculate change
        change = cashAmount - finalCost

        // If change is negative, ask for more cash
        IF change < 0 THEN
            DISPLAY "Insufficient cash. Please provide more."
            // Repeat the cash input process
        ELSE
            // Give change to customer
            DISPLAY "Your change is: " + change
            GIVE change
        END IF
    ELSE IF paymentMethod == "card" THEN
        // Process card payment (this part would involve external systems)
        DISPLAY "Please insert or swipe your card."
        // Wait for card processing to complete
        // If payment is successful:
        DISPLAY "Payment successful!"
