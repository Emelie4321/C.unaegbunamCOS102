```python
def calculate_atr():
    print("--- Izifin Technology: ATR Calculator ---")
    
    try:
        # Taking inputs from the user
        experience = float(input("Enter years of work experience: "))
        age = int(input("Enter age of the staff: "))

        # Conditional logic to determine ATR
        if experience > 25 and age >= 55:
            atr = 5600000
        elif experience > 20 and age >= 45:
            atr = 4480000
        elif experience > 10 and age >= 35:
            atr = 1500000
        else:
            # Default for staff with less than 10 years experience and below 35 years of age
            atr = 550000

        # Displaying the result
        print(f"\nThe Annual Tax Revenue (ATR) for the staff is: ₦{atr:,}")

    except ValueError:
        print("Invalid input. Please enter numerical values for age and experience.")

# Run the program
if __name__ == "__main__":
    calculate_atr()
```
