```python
# Python program to solve Annuity plan

try:
# 1. Taking all inputs
    pmt = abs(float(input("Enter fixed payment amount: ")))
    n = abs(float(input("Enter number of payment period: ")))
    t = abs(float(input("Enter the time in years: ")))
    r = abs(float(input("Enter interest rate: ")))

# 2. Computing the values (Indented to stay inside 'try') 
    if r == 0:
        a = pmt * n * t
    else:
        # Formula for Future Value of Ordinary Annuity
        a = pmt * ((1 + (r/100)/n) ** (n*t) - 1) / ((r/100)/n)

# 3. Formatting the result    
    print(f"\nThe total future value is: {a:,.2f}")

except ValueError:
    print("Error: Dear user, enter numbers only.")
except ZeroDivisionError:
    print("Error: 'Number of times' (n) cannot be zero.")


```
