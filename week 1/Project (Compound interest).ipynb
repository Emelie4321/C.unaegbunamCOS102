```python
# Python Program for calculating compound interest

# 1. Taking all inputs
try:
    p = abs(float(input("Enter Principal: ")))
    r = abs(float(input("Enter rate of interest: ")))
    t = abs(float(input("Enter time of the interest: ")))
    n = abs(float(input("Enter number of times interest is componded per year: ")))
# Computing the values of amount 
    a = p * (pow((1 + r / n), n * t))

# Calculate the compound interest
    ci = a - p

    print(f"The amount is: {a:,.2f}")
    print(f"The compound interest is: {ci:,.2f}")

except ValueError:
    print("Error: Dear user, please enter numbers only.")
    
```
